```python
import pandas as pd
import numpy as np
data= np.array([[14,35,14],[71,24,71],[11,34,19],[82,46,99,]])
df=pd.DataFrame(data,index=['Riya','Siya','Kiya','Jiya'],columns=['Python','Data_Science','DBMS',])
df
```

```python
plt.plot(df.index,df.Python,'rs:')
plt.plot(df.index,df.Data_Science,'bv-')
plt.plot(df.index,df.DBMS,'-ok')
```

```python
fig=plt.figure()#returns empty canvas
ax=fig.add_axes([.1,.1,.5,.8])
ax.plot(df.index,df.Python)
```

```python
fig=plt.figure()#returns empty canvas
axes=fig.add_axes([.3,.1,.5,.5])
axes.plot(df.index,df.Python)
```

```python
x=np.arange(0,10,2)
ay=[1,1.25,2,2.75,3]
by=[1,1,1,1,1]
cy=[2,1,2,1,2]
y=np.vstack([ay,by,cy])

fig, ax = plt.subplots()

ax.stackplot(x,y)

ax.set(xlim=(0,8), xticks=np.arange(1,8),
      ylim=(0,8), yticks=np.arange(1,8))

plt.show()
```

## Pairplot
pairplot will plot pairwise relationships across an entire dataframe
(for the numerical columns) and supports a color hue argument (for categorical columns).

```python
sns.pairplot(tips, hue='sex', palette='b')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
